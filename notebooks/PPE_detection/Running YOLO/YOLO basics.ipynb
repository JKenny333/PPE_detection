```python
import torch
from matplotlib import pyplot as plt

# Load the model
model = torch.hub.load('ultralytics/yolo/v8', 'custom', path='yolov8n.pt')  # Adjust model name as necessary
```

```python

# Load the image
img = 'test images/construction_worker.jpg'

# Perform inference
results = model(img)

# Results show
results.show()

# For Jupyter Notebooks, use the following to display the image
plt.imshow(plt.imread(img))
plt.axis('off')  # to turn off axis
plt.show()
```

```python
import torch
from PIL import Image
from yolov5 import YOLOv5  # Importing the YOLOv5 model
```

```python
model = YOLOv5("../yolo_weights/yolov5l.pt")  # Use "cuda" for GPU
```

```python
image_path = 'test images/construction_worker.jpg'
img = Image.open(image_path)
```

```python
results = model.predict(img)
```

```python
# Printing results
print(results.names)  # Print available class names
print(results.xyxy[0])  # Print detections ([xmin, ymin, xmax, ymax, confidence, class])

# For visualization in a Jupyter Notebook
results.show()
```
